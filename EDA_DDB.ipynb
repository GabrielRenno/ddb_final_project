```python
import pandas as pd
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
hotel = pd.read_csv("Hotel Reservations.csv")
```

```python
hotel
```

```python
#basic info on the data
hotel.info()
```

```python
#descriptive statistics on the data
hotel.describe()
```

```python
#check for null values
hotel.isnull().sum()
```

```python
hotel.hist(figsize=(20,15))
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x=hotel['room_type_reserved'], data=hotel)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x=hotel['type_of_meal_plan'], data=hotel)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x=hotel['market_segment_type'], data=hotel)
plt.show()
```

```python
# Plot the distribution of no_of_weekend_nights, no_of_week_nights, and total_nights
plt.hist([hotel['no_of_weekend_nights'], hotel['no_of_week_nights']], color=['orange', 'green'], label=['Weekend Nights', 'Week Nights'])
plt.title('Distribution of Nights Stayed')
plt.xlabel('No. of Nights')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
repeated_guests = hotel[hotel['repeated_guest'] == 1]
market_counts = repeated_guests['market_segment_type'].value_counts()

plt.bar(market_counts.index, market_counts.values)
plt.title('Market Segment Type for Repeated Guests')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
repeated_guests = hotel[hotel['repeated_guest'] == 0]
market_counts = repeated_guests['market_segment_type'].value_counts()

plt.bar(market_counts.index, market_counts.values)
plt.title('Market Segment Type for Non-Repeated Guests')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
# Group the data by no_of_children and calculate the mean no_of_special_requests
grouped = hotel.groupby('no_of_children')['no_of_special_requests'].mean().reset_index()

plt.bar(grouped['no_of_children'], grouped['no_of_special_requests'])
plt.title('No. of Children & No. of Special Requests')
plt.xlabel('No. of Children')
plt.ylabel('No. of Special Requests')
plt.show()
```
